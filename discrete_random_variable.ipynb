Context: Discrete Random Variable Experiment

Objective: Perform a random experiment involving the rolling of two dice.

Tasks:

Rolling the Dice: Conduct the experiment by rolling two dice.

Calculating: Determine the sum of the two outcomes.

Visualizing: Create visual representations for:

Probability Mass Function (PMF): To show the probability distribution of the sums.

Cumulative Distribution Function (CDF): To illustrate the cumulative probabilities.

Importing the libraries

```python
import random
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Roll two dice together ðŸŽ²ðŸŽ²

```python
l=[]
for i in range(36):
   dice1= random.randint(1,6)
   dice2=random.randint(1,6)
   sum=dice1+dice2
   l.append(sum)  
```

find the probability of each outcome(pmf) i.e. sum of 2 dice out of total number of possible outcome(theoretically 36 and in this experiment dice is also rolled 36 times)

```python
outcome_count=pd.Series(l).value_counts()
print(outcome_count)
outcome_probablity=outcome_count/pd.Series(l).value_counts().sum()
pmf=outcome_probablity.sort_index()
pmf
```

find out the cumulative sum(cdf)

```python
cdf=np.cumsum(pmf)
print(cdf)
```

Create the bar graph for the pmf

```python
pmf.plot(kind='bar')
```

Visualize cdf by creating the bar graph

```python
cdf.plot(kind='bar')
```
