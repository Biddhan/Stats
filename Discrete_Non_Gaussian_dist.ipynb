<h4>Importing Necessary Library</h4>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

<h4>Binomial Distribution</h4>

```python
n = 3  # number of trials
p = 0.5  # probability of success
size = 100  # number of samples

# Generate random samples from a binomial distribution
sample = np.random.binomial(n, p, size)

# Get the values for plotting the PMF
x = np.arange(0, n+1)  # possible outcomes for a binomial distribution (0 to n)
pmf_values = stats.binom.pmf(x, n, p)

# Plot the PMF
plt.stem(x, pmf_values, basefmt=" ")
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution PMF (n=3, p=0.5)')
plt.show()
```
