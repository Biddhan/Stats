<h4>Normal qq-plot(The code below is used for checking gaussian distribution(Normal distribution))</h4>

```python
import seaborn as sns
import numpy as np
dataset=sns.load_dataset('iris')
temp=dataset['sepal_length'].tolist()
y=[]
for i in range(1,101):
    y.append(np.percentile(temp,i))
temp2=np.random.normal(loc=0,scale=1,size=100)
x=[]
for i in range(1,101):
    x.append(np.percentile(temp2,i))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='red')
plt.title('QQ-plot')
plt.xlabel('Theoretical Normal distribution')
plt.ylabel('Sepal_Length')
plt.show()
```

```python
sns.scatterplot(x=x,y=y)
```

```python
import statsmodels.api as sm
fig=sm.qqplot(dataset['sepal_length'],line='45',fit=True)
```

```python
import scipy.stats as stats
stats.probplot(dataset['sepal_length'],dist="norm",plot=plt)
```

<h3>Non-Gaussian Distribution:</h3>

Uniform dist

```python
import scipy.stats as stats
x=np.random.uniform(low=0,high=1,size=100)
params=stats.uniform.fit(x)
dist=stats.uniform(loc=params[0],scale=params[1])
print(dist)
figure=sm.qqplot(x,dist=dist,line='45')
```

lognormal

```python
normaldata=np.random.normal(loc=0,scale=1,size=100)
lognormaldata=np.exp(normaldata)
print(lognormaldata)
plt.hist(lognormaldata,bins=10)
```

```python
# sns.histplot(lognormaldata)
sns.kdeplot(lognormaldata)
```

Pareto Distribution

```python
alpha = 1.16
xm = 1

# Generate sample data from a Pareto distribution
sample = stats.pareto.rvs(alpha, scale=xm, size=500)

# Fit the Pareto distribution to the sample data
params = stats.pareto.fit(sample, floc=0)
alphasample, locsample, scalesample = params

# Print estimated parameters
print(f"Estimated alpha: {alphasample}, loc: {locsample}, scale: {scalesample}")

# Plot the histogram of the sample data
plt.hist(sample, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Sample Data')

# Plot the PDF of the fitted Pareto distribution
x = np.linspace(sample.min(), 40, 500)
pdf = stats.pareto.pdf(x, alphasample, loc=locsample, scale=scalesample)
plt.plot(x, pdf, 'k', linewidth=2, label='Fitted Pareto PDF')

# Adding titles and labels
plt.title('Pareto Distribution - Sample Data vs. Fitted PDF')
plt.xlabel('X')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
log_x=np.log(x)
log_y=np.log(pdf)
plt.plot(log_x,log_y)
plt.xlabel('log(x)')
plt.ylabel('log(y)')
plt.title('log-log plot')
plt.show()
```

```python
dist=stats.pareto(b=params[0],scale=scalesample)
fig=sm.qqplot(sample,dist=dist,line='45')
plt.title('QQ plot')
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Parameters for the original Pareto distribution
alpha_true = 1.16
xm = 1

# Generate a sample from the Pareto distribution
sample = np.random.pareto(alpha_true, size=500) + xm  # Shift by xm

# Fit the Pareto distribution to the sample data
params = stats.pareto.fit(sample,floc=0)  # Fix the location parameter to 0 for a more reasonable fit

# Extract the fitted parameters
alpha_fitted, loc_fitted, scale_fitted = params

# Print the estimated parameters
print(f"True alpha: {alpha_true}")
print(f"Estimated alpha: {alpha_fitted}, loc: {loc_fitted}, scale: {scale_fitted}")
dist=stats.pareto(alpha_fitted,loc=loc_fitted,scale=scale_fitted)
fig=sm.qqplot(sample,dist=dist,line='45')
plt.show()
```
