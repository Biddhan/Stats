Importing the library

```python
import numpy as np 
from numpy.random import normal
import matplotlib.pyplot as plt
import seaborn as sns
```

Generate the samples from the normally distributed population

```python
sample=normal(loc=25,scale=12,size=500)
```

plot the histogram for the generated sample

```python
plt.hist(sample,bins=10)
```

As we have assumed the population is normally distributed calculate the pdf of the for the each sample using pdf formula for the normal distribution

```python
from scipy.stats import norm
dist=norm(sample.mean(),sample.std())
# generating some 100 evenly spaced data points from out sample
values=np.linspace(sample.min(),sample.max(),100)
probablity_density=[dist.pdf(value) for value in values]
```

Visulizing the pdf graph by plotting the sample values and its corresponding probablity density

```python
plt.hist(sample,bins=10,density=True)
plt.plot(values,probablity_density)
```

```python
sns.displot(sample)
plt.show()
sns.histplot(sample)
plt.show()
sns.kdeplot(sample)
plt.show()
```
