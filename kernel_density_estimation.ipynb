<h4>Here we are using kernel density estimation of the bimodal graph to calculate the probablity density for the population sample</h4>

Import the libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Generate the sample for the bimodal graph 

```python
sample1=np.random.normal(loc=20,scale=6,size=300)
sample2=np.random.normal(loc=40,scale=4,size=200)
sample=np.hstack((sample1,sample2))
plt.hist(sample,bins=10)
```

using gaussian as a kernerl calculate the probablity density for the bimodal data set

```python
from sklearn.neighbors import KernelDensity
model=KernelDensity(bandwidth=2,kernel='gaussian')
sample=sample.reshape((len(sample),1))
model.fit(sample)
#generate 100 evenly spaced datapoints from the sample 
values=np.linspace(sample.min(),sample.max(),100)
values=values.reshape(len(values),1)
log_probablity_density=model.score_samples(values)
probablity_density=np.exp(log_probablity_density)
```

Visualizing the pdf graph for the bimodal graph 

```python
plt.hist(sample,bins=10,density=True)
plt.plot(values[:],probablity_density)
plt.show()
```
