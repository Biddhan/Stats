Box-cox:


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Set seed for reproducibility
np.random.seed(42)

# Generate data
uniform_data = np.random.uniform(low=0, high=1, size=100)  # Uniform distribution
exponential_data = np.random.exponential(scale=1, size=100)  # Exponential distribution
binomial_data = np.random.binomial(n=10, p=0.5, size=100)  # Binomial distribution

# Create DataFrame
df = pd.DataFrame({
    'Uniform': uniform_data,
    'Exponential': exponential_data,
    'Binomial': binomial_data
})

# Plot KDE for each column
for column in df.columns:
    sns.kdeplot(df[column], label=column)

plt.legend()
plt.show()
```

Applying the Box-cox Transformation

```python
description = df.describe()
if (description < 0).any().any(): 
    print('cannot apply box-cox') 
else:
    print('box-cox can be applied')
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='box-cox')
df_transformed=pt.fit_transform(df)
print(type(df_transformed))
df_transformed = pd.DataFrame(df_transformed, columns=df.columns) 
print(type(df_transformed))
for column in df_transformed.columns:
    sns.kdeplot(df_transformed[column], label=column)
plt.legend()
plt.show()
```
